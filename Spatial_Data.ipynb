```python
import folium

mumbai_coordinates = (19.0760, 72.8777)
mumbai_map = folium.Map(location=mumbai_coordinates, zoom_start=10)

gateway_coordinates = (18.9220, 72.8347)
juhu_coordinates = (19.0970, 72.8265)
siddhivinayak_coordinates = (19.0177, 72.8300)

def add_marker(map_obj, coordinates, popup_text):
    folium.Marker(
        location=coordinates,
        popup=popup_text
    ).add_to(map_obj)

add_marker(mumbai_map, gateway_coordinates, 'Gateway of India')
add_marker(mumbai_map, juhu_coordinates, 'Juhu Beach')
add_marker(mumbai_map, siddhivinayak_coordinates, 'Siddhivinayak Temple')

mumbai_map
```
