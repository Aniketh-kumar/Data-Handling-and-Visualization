```python
!pip install geopandas

import geopandas as gpd
import matplotlib.pyplot as plt

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

ax = world.plot(figsize=(10, 6))

ax.plot(-80.191790, 25.761681, 'ro')
ax.plot(2.3522, 48.8566, 'bo')
ax.plot(77.1025, 28.7041, 'go')

ax.annotate('Miami, USA', xy=(-80.191790, 25.761681), xytext=(-60, 30),
             textcoords='offset points', arrowprops=dict(facecolor='black', arrowstyle='->'))
ax.annotate('Paris, France', xy=(2.3522, 48.8566), xytext=(-60, 30),
             textcoords='offset points', arrowprops=dict(facecolor='black', arrowstyle='->'))
ax.annotate('New Delhi, India', xy=(77.1025, 28.7041), xytext=(20, -40),
             textcoords='offset points', arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.title('World Map with Markers and Country Names')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()


```
